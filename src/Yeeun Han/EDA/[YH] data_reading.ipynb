### 가설1- 다음 변수들은 경쟁률 상승에 영향을 미칠 것이다
- 공급규모, 투기과열지구, 조정대상지역, 분양가상한제, 대규모택지개발지구, 수도권내민영공공주택지구, 공급세대수, 평면유형, 전용면적

```python
# 수치형 변수 확인
import matplotlib.pyplot as plt
import seaborn as sns

numeric_features = ["공급규모", "공급세대수", "전용면적"]

correlation = df[numeric_features + ["경쟁률"]].corr()
print(correlation["경쟁률"].sort_values(ascending=False))

plt.figure(figsize=(12, 8))
for i, col in enumerate(numeric_features, 1):
    plt.subplot(2, 2, i)
    sns.regplot(x=df[col], y=df["경쟁률"], scatter_kws={"alpha": 0.3})
    plt.title(f"{col} vs 경쟁률")

plt.tight_layout()
plt.show()
```

```python
# 범주형 변수 확인
category_features = ['투기과열지구', '조정대상지역', '분양가상한제', '대규모택지개발지구', '수도권내민영공공주택지구', '평면유형']

plt.figure(figsize=(10, 10))

for i, feature in enumerate(category_features, 1):
    plt.subplot(3, 2, i)
    sns.boxplot(x=feature, y='경쟁률', data=df)
    plt.title(f'{feature} vs 경쟁률')

plt.tight_layout()
plt.show()
```

```python
import scipy.stats as stats

for col in category_features:
    groups = [df[df[col] == val]["경쟁률"] for val in df[col].unique()]
    f_stat, p_value = stats.f_oneway(*groups)
    print(f"{col}: p-value = {p_value:.4f}")

    if p_value < 0.05:
        print(f"➡ {col}은(는) 경쟁률과 유의미한 차이가 있음")
    else:
        print(f"➡ {col}은(는) 경쟁률과 큰 차이가 없음")
```

### 가설 2: 평면유형이 다양한 매물일수록 경쟁률이 높을 것이다 

```python
# 평면유형 개수 확인
print(f"평면유형 클래스 개수: {df['평면유형'].nunique()}개")

# 평면유형별 데이터 수 확인
print(df['평면유형'].value_counts())
```

```python
# 매물별 제공된 평면 개수 계산
df['평면유형_개수'] = df.groupby('주택명')['평면유형'].transform('nunique')

plt.figure(figsize=(8, 8))
sns.boxplot(x=df['평면유형_개수'], y=df['경쟁률'])
plt.title('평면유형 개수 vs 경쟁률')
plt.xlabel('제공된 평면 개수')
plt.ylabel('경쟁률')
plt.xticks(rotation=45)
plt.show()
```

```python
# 평면 개수 10개 이상 vs 미만 단지의 경쟁률 차이 분석
import numpy as np

# 매물별 제공된 평면 개수 계산
df['평면유형_개수'] = df.groupby('주택명')['평면유형'].transform('nunique')

# 10개 이상 평면을 제공한 단지와 그렇지 않은 단지 나누기
df_10_over = df[df['평면유형_개수'] >= 10]
df_10_under = df[df['평면유형_개수'] < 10]

# 평균 경쟁률 비교
mean_10_over = np.round(df_10_over['경쟁률'].mean(), 2)
mean_10_under = np.round(df_10_under['경쟁률'].mean(), 2)

print(f'평면 10개 이상 단지 평균 경쟁률: {mean_10_over} 대 1')
print(f'평면 10개 미만 단지 평균 경쟁률: {mean_10_under} 대 1')

# 두 그룹 간 경쟁률 분포 시각화
plt.figure(figsize=(10, 8))
sns.boxplot(x=df['평면유형_개수'] >= 10, y=df['경쟁률'])
plt.xticks([0, 1], ['평면 10개 미만', '평면 10개 이상'])
plt.title('평면 개수 10개 이상 vs 미만 단지 경쟁률 비교')
plt.xlabel('단지 유형')
plt.ylabel('경쟁률')
plt.show()
```

```python
# 통계 검정(평면유형 개수 평균 차이가 유의미한지 확인)
from scipy.stats import ttest_ind

# 독립표본 t-검정 수행
t_stat, p_value = ttest_ind(df_10_over['경쟁률'], df_10_under['경쟁률'], equal_var=False)

print(f't-statistic: {t_stat:.4f}, p-value: {p_value:.4f}')

if p_value < 0.05:
    print('평면유형 개수가 경쟁률에 유의미한 영향을 미침')
else:
    print('평면유형 개수와 경쟁률 사이에 유의미한 차이가 없음')
```

### 가설3: 수도권에서는 소형 매물의 경쟁률이 높을 것이다

```python
# 수도권과 비수도권 구분
# 수도권 여부 정의(수도권: 서울, 경기, 인천)
df['수도권'] = df['공급지역명'].apply(lambda x: '수도권' if x in ['서울', '경기', '인천'] else '비수도권')

# 수도권/비수도권 데이터 개수 확인
print(df['수도권'].value_counts())
```

```python
# 전용면적 기준으로 소형(60㎡ 이하) vs 중대형(초과) 구분
df['주택유형'] = df['전용면적'].apply(lambda x: '소형' if x <= 60 else '중대형')

# 수도권 데이터만 필터링
df_seoul = df[df['수도권'] == '수도권']

# 수도권에서 소형/중대형 데이터 개수 확인
print(df_seoul['주택유형'].value_counts())
```

```python
# 수도권에서 소형과 중대형 경쟁률 비교
plt.figure(figsize=(10, 8))
sns.boxplot(x=df_seoul['주택유형'], y=df_seoul['경쟁률'])
plt.title('수도권 소형 vs 중대형 아파트 경쟁률 비교')
plt.xlabel('주택유형')
plt.ylabel('경쟁률')
plt.show()
```

```python
# 평균 경쟁률 비교(수치로 확인)
# 소형과 중대형의 평균 경쟁률 비교
mean_small = np.round(df_seoul[df_seoul['주택유형'] == '소형']['경쟁률'].mean(), 2)
mean_large = np.round(df_seoul[df_seoul['주택유형'] == '중대형']['경쟁률'].mean(), 2)

print(f'수도권 소형 아파트 평균 경쟁률: {mean_small} 대 1')
print(f'수도권 중대형 아파트 평균 경쟁률: {mean_large} 대 1')
```

```python
# 통계적 검정(t-test)

# 소형과 중대형의 경쟁률 데이터 가져오기
small_competition = df_seoul[df_seoul['주택유형'] == '소형']['경쟁률']
large_competition = df_seoul[df_seoul['주택유형'] == '중대형']['경쟁률']

# t-test 수행
t_stat, p_value = ttest_ind(small_competition, large_competition, equal_var=False)

print(f"t-statistic: {t_stat:.4f}, p-value: {p_value:.4f}")

if p_value < 0.05:
    print('수도권에서 소형 아파트의 경쟁률이 중대형보다 유의미하게 높음')
else:
    print('수도권에서 소형과 중대형 경쟁률 차이가 통계적으로 유의미하지 않음')
```

### 가설4: 공급세대수가 많을수록 경쟁률이 높을 것이다

```python
# 공급세대수와 경쟁률의 관계 분석(상관계수 확인 및 시각화)

# 공급세대수와 경쟁률 간 상관관계 분석
correlation = df[['공급세대수', '경쟁률']].corr()
print(f'공급세대수 vs 경쟁률 상관계수:\n{correlation}')

# 산점도 및 회귀선 시각화
plt.figure(figsize=(10, 8))
sns.regplot(x=df['공급세대수'], y=df['경쟁률'], scatter_kws={'alpha':0.3}, line_kws={'color':'red'})
plt.title('공급세대수 vs 경쟁률')
plt.xlabel('공급세대수')
plt.ylabel('경쟁률')
plt.show()
```

```python
# 공급세대수 구간별 경쟁률 비교(대단지 vs 소단지)
# 공급세대수를 기준으로 대단지(1000세대 이상) vs 소단지(1000세대 미만) 구분
df['단지규모'] = np.where(df['공급세대수'] >= 1000, '대단지', '소단지')

# 대단지 vs 소단지 경쟁률 비교
plt.figure(figsize=(10, 8))
sns.boxplot(x=df['단지규모'], y=df['경쟁률'])
plt.title('대단지 vs 소단지 경쟁률 비교')
plt.xlabel('단지 유형')
plt.ylabel('경쟁률')
plt.show()

# 평균 경쟁률 비교
mean_large = df[df['단지규모'] == '대단지']['경쟁률'].mean()
mean_small = df[df['단지규모'] == '소단지']['경쟁률'].mean()
print(f'대단지 평균 경쟁률: {mean_large:.2f} 대 1')
print(f'소단지 평균 경쟁률: {mean_small:.2f} 대 1')
```

```python
# 수도권 내 대단지 vs 소단지 경쟁률 분석
# 공급세대수를 기준으로 대단지(1000세대 이상) vs 소단지(1000세대 미만) 구분
df_seoul['단지규모'] = np.where(df_seoul['공급세대수'] >= 1000, '대단지', '소단지')

# 박스플롯 시각화
plt.figure(figsize=(8, 6))
sns.boxplot(x=df_seoul['단지규모'], y=df_seoul['경쟁률'])
plt.title('수도권 대단지 vs 소단지 경쟁률 비교')
plt.xlabel('단지 유형')
plt.ylabel('경쟁률')
plt.show()

# 평균 경쟁률 계산
mean_large = df_seoul[df_seoul['단지규모'] == '대단지']['경쟁률'].mean()
mean_small = df_seoul[df_seoul['단지규모'] == '소단지']['경쟁률'].mean()

print(f'수도권 대단지 평균 경쟁률: {mean_large:.2f} 대 1')
print(f'수도권 소단지 평균 경쟁률: {mean_small:.2f} 대 1')
```

```python
# 상위 1% 이상치 기준값 찾기
q99 = df_seoul['경쟁률'].quantile(0.99)

# 이상치 제거한 데이터셋
df_clean = df_seoul[df_seoul['경쟁률'] <= q99]

# 평균 경쟁률 재계산
mean_large_clean = df_clean[df_clean['단지규모'] == '대단지']['경쟁률'].mean()
mean_small_clean = df_clean[df_clean['단지규모'] == '소단지']['경쟁률'].mean()

print(f'이상치 제거 후 수도권 대단지 평균 경쟁률: {mean_large_clean:.2f} 대 1')
print(f'이상치 제거 후 수도권 소단지 평균 경쟁률: {mean_small_clean:.2f} 대 1')

# 이상치 제거 후 박스플롯 다시 그리기
plt.figure(figsize=(8, 6))
sns.boxplot(x=df_clean['단지규모'], y=df_clean['경쟁률'])
plt.title('이상치 제거 후 수도권 대단지 vs 소단지 경쟁률 비교')
plt.xlabel('단지 유형')
plt.ylabel('1순위 청약 경쟁률')
plt.show()
```

```python
# 대단지 vs 소단지 경쟁률 차이가 통계적으로 유의미한지 검정
from scipy.stats import ttest_ind

# 독립표본 t-검정 수행(등분산 가정 X)
t_stat, p_value = ttest_ind(df_clean[df_clean['단지규모'] == '대단지']['경쟁률'],
                            df_clean[df_clean['단지규모'] == '소단지']['경쟁률'],
                            equal_var=False)

print(f't-statistic: {t_stat:.4f}, p-value: {p_value:.4f}')

if p_value < 0.05:
    print('수도권 대단지 vs 소단지 경쟁률 차이는 통계적으로 유의미함')
else:
    print('수도권 대단지 vs 소단지 경쟁률 차이는 통계적으로 유의미하지 않음')
```
