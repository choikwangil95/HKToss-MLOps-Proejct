```python
import pandas as pd

df = pd.read_csv("../storage/train_data/train-250312-01.csv", encoding="cp949")
df
```

```python
df.columns
```

```python
df['공급금액(최고가 기준)'] = df['공급금액(최고가 기준)'] / 10000
df['공급금액(최고가 기준)']
```

```python
# 평당가격 칼럼 추가
df['평당가격'] = df['공급금액(최고가 기준)'] / df['전용면적']
df['평당가격']
```

```python
dummy = pd.read_csv("C:/Users/user/mid-project/src/storage/raw_data/병합_청약매물_목록_정보 (서울, 경기, 인천)_위도경도추가_2.csv", encoding='cp949')

dummy.columns
```

```python
dummy.head()
```

```python
# 계약연도/계약월 칼럼 추가
dummy['계약연도'] = pd.to_datetime(dummy['청약접수종료일']).dt.year
dummy['계약월'] = pd.to_datetime(dummy['청약접수종료일']).dt.month
```

```python
dummy = dummy[["공고번호", "주택명", "계약연도", "계약월", "공급지역명", "전용면적", "시군구", "읍면동"]]
dummy.head()
```

```python
dummy
```

```python
dummy.drop_duplicates(subset=["공고번호"], keep='first')
```

```python
df = df[['공고번호', '공급금액(최고가 기준)', '평당가격']]
df.head()
```

```python
data = pd.merge(df, dummy.drop_duplicates(subset=["공고번호"], keep='first'), on='공고번호', how='inner')
data
```

```python
# 칼럼명 변경 (공급금액_평당가격으로)
data.rename(columns={'평당가격': '공급금액_평당가격'}, inplace=True)
```

```python
# 지역 동 별 평당가격
data.to_csv("./공급금액_평당가격.csv", index=False, encoding='cp949')
data
```
