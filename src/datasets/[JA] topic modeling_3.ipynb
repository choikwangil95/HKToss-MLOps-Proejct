```python
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from sklearn.decomposition import LatentDirichletAllocation
from konlpy.tag import Okt
import re
```

## 1. 전처리

```python
# 불용어 txt 파일 리스트로 변환
stopwords_file = '/Users/JooAnLee/mid_project/src/datasets/stopwords-ko.txt'

def load_stopwords(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        # 파일에서 각 줄을 읽어서 리스트로 반환 (strip()으로 공백 제거)
        stopwords = [line.strip() for line in file.readlines()]
    return stopwords

stopwords = load_stopwords(stopwords_file)
print(stopwords[:10])
```

```python
# 코퍼스 생성 (content 컬럼 데이터를 리스트로 변환)
corpus = df['content'].tolist()
print(corpus[:3])
```

```python
# 텍스트 정제 함수
def clean_text(text):
    # 1. 괄호 (), [] 안의 내용 제거
    text = re.sub(r'\[.*?\]|\(.*?\)', '', text) 
    # 2. 한글과 공백 제외한 문자 제거
    text = re.sub(r'[^가-힣\s]', '', text) 
    # 3. 연속된 공백을 하나의 공백으로 변경
    text = re.sub(r'\s+', ' ', text)
    # 4. 문장 앞뒤 공백 제거
    text = text.strip()
    return text

cleaned_corpus = [clean_text(text) for text in corpus]
print(cleaned_corpus[:3])
```

```python
okt = Okt()

# 전처리 함수 (명사 추출 + 불용어 제거 + 짧은 단어 제거)
def preprocess(text, stopwords, min_length=2):
    # 1. 텍스트 정제
    text = clean_text(text)  # 괄호 내용 제거, 특수문자 제거 등

    # 2. 명사 추출
    nouns = okt.nouns(text)
    
    # 3. 불용어 제거 및 짧은 단어(길이 2 이하) 제거
    filtered_nouns = [word for word in nouns if word not in stopwords and len(word) > min_length]
    
    # 4. 공백으로 구분된 문자열로 결합
    return ' '.join(filtered_nouns)

# 전처리된 코퍼스 생성
processed_corpus = [preprocess(text, stopwords) for text in cleaned_corpus]

# 전처리된 코퍼스 확인
print(processed_corpus[:5])
```

```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
from konlpy.tag import Okt
import re
from sklearn.metrics import log_loss

# 1. 벡터화 (DTM 생성)
vectorizer = CountVectorizer()
dtm = vectorizer.fit_transform(processed_corpus)
print("DTM shape:", dtm.shape)

# 2. 최적 토픽 개수 찾기 (Perplexity 활용)
def calculate_perplexity(dtm, num_topics):
    lda_model = LatentDirichletAllocation(n_components=num_topics, random_state=42)
    lda_model.fit(dtm)
    return lda_model.perplexity(dtm)

topic_nums = range(2, 15) # 일반적으로 2개부터 시작하여 적절한 범위 설정
perplexities = [calculate_perplexity(dtm, num_topics) for num_topics in topic_nums]

import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.plot(topic_nums, perplexities, marker='o')
plt.title('Perplexity for Different Number of Topics')
plt.xlabel('Number of Topics')
plt.ylabel('Perplexity')
plt.grid(True)
plt.show()

# Perplexity가 가장 낮아지는 지점을 시각적으로 확인하고, 최적의 토픽 개수를 선택합니다.
# 실제로는 elbow point (급격하게 감소하다 완만해지는 지점)를 기준으로 선택하는 경우가 많습니다.
# 여기서는 예시로 perplexity가 가장 낮은 토픽 개수를 선택합니다.
optimal_topic_num = topic_nums[np.argmin(perplexities)]
print(f"Optimal number of topics based on perplexity: {optimal_topic_num}")

# 3. LDA 모델 학습 (최적 토픽 개수로)
lda_model = LatentDirichletAllocation(n_components=optimal_topic_num, random_state=42)
lda_model.fit(dtm)

# 4. 성능 평가 (Perplexity 값 출력)
print(f"Perplexity of the LDA model with {optimal_topic_num} topics: {lda_model.perplexity(dtm):.4f}")

# 추가적인 성능 평가 지표를 원하시면 다른 방법을 고려해야 합니다. (예: Topic Coherence)

# 5. 토픽 이름 정하기
feature_names = vectorizer.get_feature_names_out()

def display_topics(model, feature_names, num_top_words=10):
    topics = []
    for topic_idx, topic in enumerate(model.components_):
        top_words_indices = topic.argsort()[:-num_top_words - 1:-1]
        top_words = [feature_names[i] for i in top_words_indices]
        topics.append(top_words)
        print(f"Topic #{topic_idx + 1}: {', '.join(top_words)}")
    return topics

topic_keywords = display_topics(lda_model, feature_names)

# 만들어진 군집들로 토픽 이름 정하기 (수동으로 이름을 부여해야 합니다.)
topic_names = {}
for i, keywords in enumerate(topic_keywords):
    print(f"\nTopic #{i + 1} 키워드: {keywords}")
    topic_name = input(f"Topic #{i + 1}의 이름을 정해주세요: ")
    topic_names[f"Topic {i + 1}"] = topic_name

print("\n정해진 토픽 이름:")
for topic, name in topic_names.items():
    print(f"{topic}: {name}")

# (선택 사항) 각 문서에 할당된 토픽 확인
doc_topic_dist = lda_model.transform(dtm)
df['topic'] = np.argmax(doc_topic_dist, axis=1) + 1 # 토픽 번호는 1부터 시작

print("\n각 문서에 할당된 토픽 (상위 10개):")
print(df[['title', 'topic']].head(10))
```

```python
import matplotlib.pyplot as plt
import pyLDAvis
import pyLDAvis.sklearn
from wordcloud import WordCloud
from tqdm import tqdm
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
from sklearn.model_selection import GridSearchCV
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

# 1. DTM(Document-Term Matrix) 생성
print("DTM 생성 중...")
vectorizer = CountVectorizer(max_df=0.9, min_df=2)  # 최대 90%의 문서에 등장하는 단어 제외, 최소 2개 이상의 문서에 등장하는 단어만 포함
dtm = vectorizer.fit_transform(processed_corpus)
print(f"DTM 크기: {dtm.shape}")

# 단어 사전 생성 (단어: 인덱스)
feature_names = vectorizer.get_feature_names_out()
print(f"총 단어 수: {len(feature_names)}")
print(f"단어 사전 일부: {feature_names[:10]}")

# 2. 최적의 토픽 개수 찾기 - Perplexity 사용
def compute_perplexity_values(dtm, topic_range):
    perplexities = []
    likelihoods = []
    
    for n_topic in tqdm(topic_range):
        lda = LatentDirichletAllocation(
            n_components=n_topic,
            learning_method='online',
            random_state=42,
            max_iter=10
        )
        lda.fit(dtm)
        perplexities.append(lda.perplexity(dtm))
        likelihoods.append(lda.score(dtm))
    
    return perplexities, likelihoods

print("최적의 토픽 개수 탐색 중...")
topic_range = range(5, 31, 5)  # 5부터 30까지 5 단위로 탐색
perplexities, likelihoods = compute_perplexity_values(dtm, topic_range)

# 그래프로 시각화
plt.figure(figsize=(12, 6))

# Perplexity 그래프 
plt.subplot(1, 2, 1)
plt.plot(topic_range, perplexities, marker='o')
plt.title('Perplexity by Number of Topics')
plt.xlabel('Number of Topics')
plt.ylabel('Perplexity')

# Log-Likelihood 그래프
plt.subplot(1, 2, 2)
plt.plot(topic_range, likelihoods, marker='o')
plt.title('Log-Likelihood by Number of Topics')
plt.xlabel('Number of Topics')
plt.ylabel('Log-Likelihood')

plt.tight_layout()
plt.savefig('topic_number_evaluation.png')
plt.show()

# 3. 최적의 하이퍼파라미터 찾기 (Grid Search)
print("최적의 하이퍼파라미터 탐색 중...")
# 그래프를 통해 확인한 최적의 토픽 개수 (예시로 15 설정)
optimal_n_topics = 15  # 그래프 결과에 따라 변경해야 함

# 탐색할 하이퍼파라미터 
param_grid = {
    'n_components': [optimal_n_topics],
    'learning_decay': [0.5, 0.7, 0.9],
    'max_iter': [10, 20],
    'doc_topic_prior': [0.1, 0.5, 1.0],
    'topic_word_prior': [0.01, 0.1, 0.5]
}

# Grid Search 실행
lda = LatentDirichletAllocation(random_state=42, learning_method='online')
grid_search = GridSearchCV(lda, param_grid=param_grid, cv=5, n_jobs=-1)
grid_search.fit(dtm)

# 최적의 파라미터 및 모델
best_lda_params = grid_search.best_params_
best_lda_model = grid_search.best_estimator_

print(f"최적의 파라미터: {best_lda_params}")
print(f"최적의 Perplexity: {best_lda_model.perplexity(dtm)}")

# 4. 최종 LDA 모델 학습
print("최종 LDA 모델 학습 중...")
# 최적 파라미터를 사용하여 LDA 모델 학습
final_lda = LatentDirichletAllocation(
    n_components=best_lda_params['n_components'],
    doc_topic_prior=best_lda_params['doc_topic_prior'],
    topic_word_prior=best_lda_params['topic_word_prior'],
    learning_decay=best_lda_params['learning_decay'],
    max_iter=best_lda_params['max_iter'],
    learning_method='online',
    random_state=42
)
final_lda.fit(dtm)

# 5. 토픽별 단어 분포 확인 및 시각화
def display_topics(model, feature_names, n_top_words):
    topics = {}
    for topic_idx, topic in enumerate(model.components_):
        top_features_indices = topic.argsort()[:-n_top_words-1:-1]
        top_features = [feature_names[i] for i in top_features_indices]
        topics[topic_idx] = top_features
        print(f"토픽 #{topic_idx+1}: {', '.join(top_features)}")
    return topics

# 토픽별 상위 15개 단어 확인
n_top_words = 15
print(f"\n각 토픽별 상위 {n_top_words}개 단어:")
topics = display_topics(final_lda, feature_names, n_top_words)

# 6. 각 토픽에 대한 워드클라우드 생성
print("\n토픽별 워드클라우드 생성 중...")
def generate_wordcloud(model, feature_names, topic_idx):
    topic_weights = model.components_[topic_idx]
    word_weights = {feature_names[i]: topic_weights[i] for i in range(len(feature_names))}
    
    # 가중치가 0인 단어 제외
    word_weights = {word: weight for word, weight in word_weights.items() if weight > 0}
    
    wordcloud = WordCloud(
        width=800, height=400,
        background_color='white',
        font_path='/Users/JooAnLee/Library/Fonts/NanumGothic.ttf',  # 한글 폰트 경로 (필요시 변경)
        max_words=100
    ).generate_from_frequencies(word_weights)
    
    plt.figure(figsize=(10, 6))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title(f'토픽 #{topic_idx+1} 워드클라우드')
    plt.savefig(f'topic_{topic_idx+1}_wordcloud.png')
    plt.close()

# 모든 토픽에 대한 워드클라우드 생성
for topic_idx in range(best_lda_params['n_components']):
    generate_wordcloud(final_lda, feature_names, topic_idx)

# 7. 문서별 토픽 분포 확인
print("\n문서별 토픽 분포 분석 중...")
# 각 문서에 대한 토픽 분포
doc_topic_dist = final_lda.transform(dtm)

# 각 문서의 주요 토픽 할당
dominant_topics = np.argmax(doc_topic_dist, axis=1)
df['dominant_topic'] = dominant_topics + 1  # 토픽 번호를 1부터 시작하도록 설정

# 토픽별 문서 수 확인
topic_counts = df['dominant_topic'].value_counts().sort_index()
print("\n각 토픽별 문서 수:")
print(topic_counts)

# 토픽 분포 시각화
plt.figure(figsize=(12, 6))
topic_counts.plot(kind='bar')
plt.title('Documents per Topic')
plt.xlabel('Topic ID')
plt.ylabel('Number of Documents')
plt.xticks(rotation=0)
plt.savefig('topic_distribution.png')
plt.show()

# 8. 토픽 시각화 (pyLDAvis)
print("\nLDA 토픽 시각화 (pyLDAvis) 생성 중...")
panel = pyLDAvis.sklearn.prepare(final_lda, dtm, vectorizer)
pyLDAvis.save_html(panel, 'lda_visualization.html')
print("pyLDAvis 시각화가 'lda_visualization.html' 파일로 저장되었습니다.")

# 9. 토픽 명명하기 (토픽 이름 정하기)
print("\n토픽 명명하기:")
topic_names = {}

for topic_idx, top_words in topics.items():
    print(f"\n토픽 #{topic_idx+1} 단어: {', '.join(top_words)}")
    # 여기서는 자동으로 토픽 이름을 제안
    # 실제로는 전문가가 단어 목록을 보고 적절한 이름을 정해야 함
    
    # 간단한 토픽 이름 예시 (실제 프로젝트에서는 더 적절한 이름으로 수정 필요)
    suggested_name = f"토픽{topic_idx+1}: {top_words[0]}_{top_words[1]}"
    topic_names[topic_idx+1] = suggested_name
    print(f"제안된 토픽 이름: {suggested_name}")

# 10. 결과 저장
# 토픽 별 상위 단어 저장
topic_words_df = pd.DataFrame({topic+1: words for topic, words in topics.items()})
topic_words_df.to_csv('topic_words.csv', encoding='utf-8-sig')

# 문서별 토픽 분포 및 주요 토픽 저장
df_with_topics = df.copy()
for topic_idx in range(best_lda_params['n_components']):
    df_with_topics[f'Topic_{topic_idx+1}'] = doc_topic_dist[:, topic_idx]

# 토픽 이름을 반영한 주요 토픽 열 추가
df_with_topics['dominant_topic_name'] = df_with_topics['dominant_topic'].map(topic_names)
df_with_topics.to_csv('documents_with_topics.csv', encoding='utf-8-sig')

print("\n토픽 모델링 완료! 결과가 CSV 파일로 저장되었습니다.")

# 11. 모델 성능 평가
print("\n모델 성능 평가:")
# 일관성 점수 확인 (토픽 응집도)
topic_coherence = final_lda.score(dtm)
print(f"모델 일관성 점수 (Log-likelihood): {topic_coherence:.4f}")
print(f"모델 복잡도 (Perplexity): {final_lda.perplexity(dtm):.4f}")

# 토픽 간 유사도 확인
def calculate_topic_similarity(model):
    components = model.components_ / model.components_.sum(axis=1)[:, np.newaxis]
    similarity_matrix = np.zeros((components.shape[0], components.shape[0]))
    
    for i in range(components.shape[0]):
        for j in range(components.shape[0]):
            # Jensen-Shannon 유사도
            p = components[i]
            q = components[j]
            m = 0.5 * (p + q)
            similarity_matrix[i, j] = 0.5 * (np.sum(p * np.log(p / m)) + np.sum(q * np.log(q / m)))
    
    return similarity_matrix

similarity_matrix = calculate_topic_similarity(final_lda)

# 히트맵으로 토픽 간 유사도 시각화
plt.figure(figsize=(10, 8))
plt.imshow(similarity_matrix, cmap='viridis')
plt.colorbar()
plt.title('Topic Similarity Matrix')
plt.xlabel('Topic ID')
plt.ylabel('Topic ID')
plt.savefig('topic_similarity.png')
plt.show()

print("\n토픽 모델링 프로세스가 완료되었습니다!")
```
