```python
train_article = """
의왕 월암지구 대방 디에트르 레이크파크 조감도. [사진 출처 = 대방건설]경기도 의왕시에서 분양된 단지에서 올해 첫 청약통장 만점 당첨자가 나왔다. 만점인 84점을 받으려면 7인 가구 이상으로 무주택 기간 15년 이상을 유지해야 한다. 13일 청약홈에 따르면 지난 8일 당첨자를 발표한 ‘의왕 월암지구 대방 디에트르 레이크파크’에서 84점 만점 통장이 나왔다. 만점 통장은 전용 84㎡A 기타 지역에서 나왔다. 최저 47점, 최고 84점을 기록했다. 이 단지는 앞서 진행된 1·2순위 청약에서 386가구 공급(특별공급 제외)에 1625명이 신청했다. 분양가 상한제가 적용돼 주변 시세보다 공급가격이 낮게 책정된 점이 수요자들의 관심을 끌었다. 단지 분양가는 전용 75㎡ 기준 5억4000만원대, 전용 84㎡ 기준 6억1000만원대부터 시작된다. 국토교통부 실거래가 공개시스템에 따르면 인근 위치한 ‘의왕역 푸르지오 라포레’의 전용 84㎡D타입은 지난해 7월 7억5000만원에 거래됐다. 한편 지난해에는 총 7개 단지에서 만점 통장이 나왔다. 지역별로 보면 서울 4개, 경기 2개, 전북 1개 등이다. 서울에서는 ‘래미안 원베일리’, ‘래미안 원펜타스’, ‘디에이치 대치 에델루이’ 등 분상제가 적용돼 분양가가 시세보다 저렴하게 나왔던 강남3구 아파트에 집중됐다. 올해는 이른바 ‘로또 청약’으로 불리는 분상제 적용 물량이 줄어들 것으로 예상되면서 고가점 통장 경쟁도 더 치열할 것으로 예상된다.
"""
```

### 1. 텍스트 전처리

```python
import re
import pandas as pd
from soynlp.noun import LRNounExtractor_v2
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
from konlpy.tag import Okt
```

```python
# 1. 특수문자 및 [사진 출처 = ...] 제거
train_article_cleaned = re.sub(r'\[사진 출처.*?\]', '', train_article)
train_article_cleaned = re.sub(r'[^가-힣a-zA-Z0-9\s]', '', train_article_cleaned)
```

```python
# 2. 고유명사 자동 추출 (soynlp 사용)
noun_extractor = LRNounExtractor_v2(verbose=True)
nouns = noun_extractor.train_extract([train_article_cleaned])

# 상위 고유명사 필터링 (빈도 기반)
proper_nouns = [noun for noun, score in nouns.items() if score.frequency >= 1.0]
print(f"추출된 고유명사: {proper_nouns}")

# 3. 고유명사 결합 처리 (공백 제거 → _로 결합)
for noun in proper_nouns:
    train_article_cleaned = train_article_cleaned.replace(noun, noun.replace(' ', '_'))

# 4. 형태소 분석 및 명사 추출 (Okt 사용)
okt = Okt()
tokens = okt.nouns(train_article_cleaned)

# 불용어 제거 및 한 글자 단어 제거
tokens = [token for token in tokens if len(token) > 1]

# 전처리 완료된 텍스트 확인
cleaned_article = ' '.join(tokens)
print(f"\n전처리 완료 텍스트:\n{cleaned_article}")
```

### 2. 벡터화 (DTM)

```python
# 5. DTM 생성 (CountVectorizer 사용)
vectorizer = CountVectorizer(
    max_df=1.0,  # 일단 단일문서라 1로 설정
    min_df=1,   # 최소 1회 이상 등장한 단어만 포함 (일단 단일문서라서)
    # ngram_range=(1, 2),   # unigram + bigram
    stop_words=['에서', '그리고', '하지만', '때문에', '지난', '있다', '하는', '했다', '수']
)

dtm = vectorizer.fit_transform([cleaned_article])
```

### 3. LDA 모델 학습

```python
# 6. LDA 모델 학습
lda = LatentDirichletAllocation(
    n_components=7, 
    random_state=42
)
lda.fit(dtm)

# 7. 토픽별 주요 키워드 출력
terms = vectorizer.get_feature_names_out()
print("\n[토픽별 주요 키워드]")
for i, topic in enumerate(lda.components_):
    top_keywords = [terms[j] for j in topic.argsort()[-10:]]
    print(f"토픽 {i + 1}: {top_keywords}")
```

```python
# 8. 기사별 주요 토픽 점수 출력
doc_topic = lda.transform(dtm)
topic_names = ['교통', '시세차익', '입지', '경쟁률', '개발_호재', '대출', '공시가']

print("\n[기사별 토픽 비율]")
for i, score in enumerate(doc_topic[0]):
    print(f"{topic_names[i]}: {score:.4f}")

# 9. 주요 토픽 매핑
top_topic = doc_topic[0].argmax()
print(f"\n이 기사의 주요 토픽: {topic_names[top_topic]}")
```

## 매물 하나 지정해서 모든 기사내용을 하나로 합쳐서 시도

```python
# apartment 값이 같은 경우 content 내용 합치기
#combined_contents = df.groupby('apartment')['content'].apply(' '.join).reset_index()

# 결과 확인
#combined_contents.head()
```

```python
# 0번째 아파트부터 시작
target_apartment = df.loc[0, 'apartment']

# 같은 apartment 값을 가진 content를 하나로 합치기
combined_content = ' '.join(df.loc[df['apartment'] == target_apartment, 'content'])

# 결과 확인
print(combined_content)
```

```python
# 텍스트 전처리

# 1. 특수문자 및 [사진 출처 = ...] 제거
combined_content_clean = re.sub(r'\[사진 출처.*?\]', '', combined_content)
combined_content_clean = re.sub(r'[^가-힣a-zA-Z0-9\s]', '', combined_content_clean)

# 2. 고유명사 자동 추출 (soynlp 사용)
noun_extractor = LRNounExtractor_v2(verbose=True)
nouns = noun_extractor.train_extract([combined_content_clean])

# 상위 고유명사 필터링 (빈도 기반)
proper_nouns = [noun for noun, score in nouns.items() if score.frequency >= 1.0]
print(f"추출된 고유명사: {proper_nouns}")

# 3. 고유명사 결합 처리 (공백 제거 → _로 결합)
for noun in proper_nouns:
    combined_content_clean = combined_content_clean.replace(noun, noun.replace(' ', '_'))

# 4. 형태소 분석 및 명사 추출 (Okt 사용)
okt = Okt()
tokens = okt.nouns(combined_content_clean)

# 불용어 제거 및 한 글자 단어 제거
tokens = [token for token in tokens if len(token) > 1]

# 전처리 완료된 텍스트 확인
cleaned_combined_content = ' '.join(tokens)
print(f"\n전처리 완료 텍스트:\n{cleaned_combined_content}")
```

```python
# 벡터화 - DTM 생성 (CountVectorizer 사용)
vectorizer = CountVectorizer(
    max_df=1.0,  # 일단 단일문서라 1로 설정
    min_df=1,   # 단일문서라 이것도 1....
    # ngram_range=(1, 2),   # unigram + bigram
    stop_words=['에서', '그리고', '하지만', '때문에', '지난', '있다', '하는', '했다', '수']
)

dtm2 = vectorizer.fit_transform([cleaned_combined_content])
```

```python
# LDA 모델 학습
lda = LatentDirichletAllocation(
    n_components=7, 
    random_state=42
)
lda.fit(dtm2)

# 토픽별 주요 키워드 출력
#terms2 = vectorizer.get_feature_names_out()
#print("\n[토픽별 주요 키워드]")
#for i, topic in enumerate(lda.components_):
#    top_keywords = [terms[j] for j in topic.argsort()[-10:]]
#    print(f"토픽 {i + 1}: {top_keywords}")
```

```python
terms2 = vectorizer.get_feature_names_out()
print("\n[토픽별 주요 키워드]")
for i, topic in enumerate(lda.components_):
    top_n = min(10, len(terms2))  # 피처 개수에 따라 조정
    top_keywords = [terms2[j] for j in topic.argsort()[-top_n:]]
    print(f"토픽 {i + 1}: {top_keywords}")
```

```python
# 기사별 주요 토픽 점수 출력
doc_topic = lda.transform(dtm2)
topic_names = ['교통', '시세차익', '입지', '경쟁률', '개발_호재', '대출', '공시가']

print("\n[기사별 토픽 비율]")
for i, score in enumerate(doc_topic[0]):
    print(f"{topic_names[i]}: {score:.4f}")

# 9. 주요 토픽 매핑
top_topic = doc_topic[0].argmax()
print(f"\n이 기사의 주요 토픽: {topic_names[top_topic]}")
```

일단 LDA는 문서 집합이 2개 이상일 떄 의미있는결과를 도출 가능.\
따라서 개별 기사를 각각 토픽모델링(각각 DTM생성하고 LDA 모델링) 하고
- 각 아파트 매물에 속하는 모든 기사의 토픽 분포를 가져와서
- 각 토픽별로 해당 아파트 매물 기사들의 토픽 확률을 평균내거나 합산하여 매물과 각 토픽간의 연관도 값을 계산
