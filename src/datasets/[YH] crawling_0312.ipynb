```python
# BeautifulSoup 파싱
soup = bs(html, 'html.parser')

# 모든 <a> 태그 중 'cru' 속성이 있는 것 찾기
articles = soup.find_all(attrs={"cru": True})

# User-Agent 설정 (네이버 차단 방지)
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
}

# 기사 크롤링
for article in articles:
    article_url = article['cru']
    print(f"기사 URL: {article_url}")
    
    try:
        # HTTP 요청
        response = requests.get(article_url, headers=headers)
        response.raise_for_status()  # 에러 발생 시 예외 처리

        # HTML 파싱
        article_soup = bs(response.text, 'html.parser')

        # 기사 본문 찾기 (네이버 뉴스는 'article' 태그를 사용)
        article_body = article_soup.find("article")

        if article_body:
            print("기사 내용:")
            print(article_body.get_text(strip=True))  # 기사 본문 출력
        else:
            print("기사 본문을 찾을 수 없습니다.")

    except requests.exceptions.RequestException as e:
        print(f"요청 실패: {e}")

print("크롤링 완료")
```

```python
import urllib.parse

def get_naver_news_url(apartment_name, additional_keyword):
    base_url = "https://search.naver.com/search.naver"
    query = f'"{apartment_name} {additional_keyword}"'  # 아파트 이름과 추가 키워드를 함께 검색
    
    params = {
        "ssc": "tab.news.all",
        "where": "news",
        "sm": "tab_jum",
        "query": query
    }
    
    search_url = base_url + "?" + urllib.parse.urlencode(params, quote_via=urllib.parse.quote)
    return search_url

# 예제 실행
apartment_name = "래미안 원베일리"
additional_keyword = "청약" 
news_url = get_naver_news_url(apartment_name, additional_keyword)
print(news_url)
```

```python
from bs4 import BeautifulSoup
import requests

def crawl_naver_news(keyword, max_articles=10):
    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36"
    }
    
    # 검색 URL 생성
    search_url = f"https://search.naver.com/search.naver?where=news&query={urllib.parse.quote(keyword)}"
    response = requests.get(search_url, headers=headers)
    soup = BeautifulSoup(response.text, 'html.parser')
    
    # cru 속성을 가진 기사 추출 (최대 10개)
    articles = soup.find_all(attrs={"cru": True})[:max_articles]
    results = []
    
    for idx, article in enumerate(articles, 1):
        try:
            article_url = article['cru']
            article_res = requests.get(article_url, headers=headers, timeout=5)
            article_soup = BeautifulSoup(article_res.text, 'html.parser')
            
            # 최신 HTML 구조 반영
            title = article_soup.select_one('#title_area').get_text(strip=True)
            content = article_soup.select_one('#newsct_article').get_text(strip=True)
            
            results.append({
                "no": idx,
                "title": title,
                "content": content[:200] + "..."  # 200자 요약
            })
            
            print(f"{idx}번 기사 처리 완료: {title}")
            
        except Exception as e:
            print(f"기사 처리 실패 ({article_url}): {str(e)}")
        
        if len(results) >= max_articles:
            break
            
    return results

# 실행 예시
if __name__ == "__main__":
    news_data = crawl_naver_news("래미안 원베일리 청약")
    
    print("\n[최종 결과]")
    for article in news_data:
        print(f"\n{article['no']}번 기사: {article['title']}")
        print(f"요약 내용: {article['content']}")
```
