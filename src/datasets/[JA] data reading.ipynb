## 가설
- 평균 당첨 가점이 높을수록 경쟁률이 높고
- 평균 당첨 가점과 경쟁률이 높은 곳은 서울, 그 중에서도 특정 지역구
- 경기도중에선 어디일까

```python
# 수도권만 잘라보기
capital_area = train0311[train0311['공급지역명'].isin(['서울', '경기', '인천'])]
capital_area.head()
```

```python
capital_area.info()
```

```python
# 수도권 한정, 평균당첨가점, 최고당첨가점과 경쟁률간의 상관관계
plt.figure(figsize=(12, 6))

# 평균당첨가점과 경쟁률의 상관관계
plt.subplot(1, 2, 1)
sns.scatterplot(x='평균당첨가점', y='경쟁률', data=capital_area)
plt.title('평균당첨가점과 경쟁률의 상관관계')
plt.xlabel('평균당첨가점')
plt.ylabel('경쟁률')

# 최고당첨가점과 경쟁률의 상관관계
plt.subplot(1, 2, 2)
sns.scatterplot(x='최고당첨가점', y='경쟁률', data=capital_area)
plt.title('최고당첨가점과 경쟁률의 상관관계')
plt.xlabel('최고당첨가점')
plt.ylabel('경쟁률')

plt.tight_layout()
plt.show()
```

```python
# 상관계수 계산
correlation_avg = capital_area[['평균당첨가점', '경쟁률']].corr()
correlation_max = capital_area[['최고당첨가점', '경쟁률']].corr()

# 결과 출력
print("평균당첨가점과 경쟁률의 상관계수:")
print(correlation_avg)
print("\n최고당첨가점과 경쟁률의 상관계수:")
print(correlation_max)
```

```python
# 경쟁률의 기술 통계 확인
competition_stats = capital_area['경쟁률'].describe()
print("경쟁률의 기술 통계:")
print(competition_stats)
```

```python
# 경쟁률의 75%와 max 값 사이의 데이터 개수 구하기
q75 = capital_area['경쟁률'].quantile(0.75)  # 75% 값
max_value = capital_area['경쟁률'].max()    # max 값

# 75%와 max 값 사이의 값들 필터링
filtered_data = capital_area[(train0311['경쟁률'] > q75) & (capital_area['경쟁률'] <= max_value)]

# 해당 구간에 속하는 데이터의 개수
count_filtered_data = filtered_data.shape[0]
print(f"75%에서 max 구간에 속하는 데이터의 개수: {count_filtered_data}")
```

```python
# 경쟁률이 1 이상이고, 최대값 이하인 데이터 필터링
comp_more1_data = capital_area[(capital_area['경쟁률'] >= 1) & (capital_area['경쟁률'] <= max_value)]

# 해당 구간에 속하는 데이터의 개수
count_more1_data = comp_more1_data.shape[0]
print(f"경쟁률이 1 이상, max 이하 구간에 속하는 데이터의 개수: {count_more1_data}")
```

```python
# 경쟁률이 1이상인 수도권 데이터에서 서울만 따로 보고싶음
seoul_compmore1 = comp_more1_data[comp_more1_data['공급지역명'].isin(['서울'])]
seoul_compmore1.head()
```

```python
# 서울_구와 서울_동 컬럼 삭제
comp_more1_data = comp_more1_data.drop(['서울_구', '서울_동'], axis=1)

# 결과 확인
print(comp_more1_data.head())
```

```python
# 서울_구 컬럼 생성 (구 부분 추출)
seoul_compmore1['서울_구'] = seoul_compmore1['공급위치'].str.extract(r'서울특별시\s([가-힣]+구)')

# 서울_동 컬럼 생성 (동 부분 추출)
seoul_compmore1['서울_동'] = seoul_compmore1['공급위치'].str.extract(r'([가-힣]+동)')

# 결과 확인
print(seoul_compmore1[['공급위치', '서울_구', '서울_동']].head())
```

```python
# 경쟁률 1 이상의 서울 데이터에서 서울_구별 평균 경쟁률 계산
import matplotlib.pyplot as plt

avg_seoul_compmore1 = seoul_compmore1.groupby('서울_구')['경쟁률'].mean().sort_values(ascending=False)

# 3. 결과 출력 (수치로 확인)
print("서울_구별 평균 경쟁률:")
print(avg_seoul_compmore1)

```

```python
# 시각화 (막대 그래프)
plt.figure(figsize=(12, 8))
avg_seoul_compmore1.plot(kind='bar', color='skyblue')
plt.title('서울_구별 평균 경쟁률(경쟁률 1 이상)')
plt.xlabel('서울_구')
plt.ylabel('평균 경쟁률')
plt.xticks(rotation=90)
plt.show()
```

```python
import matplotlib.pyplot as plt

# 1. 서초구 데이터 필터링
seocho_data = seoul_compmore1[seoul_compmore1['서울_구'] == '서초구']

# 2. 서초구 내 서울_동별 평균 경쟁률 계산
avg_seocho_comp = seocho_data.groupby('서울_동')['경쟁률'].mean().sort_values(ascending=False)

# 3. 결과 출력 (수치로 확인)
print("서초구 서울_동별 평균 경쟁률:")
print(avg_seocho_comp)

# 4. 서초구 서울_동별 평균 경쟁률 시각화 (막대 그래프)
plt.figure(figsize=(12, 8))
avg_seocho_comp.plot(kind='bar', color='skyblue')
plt.title('서초구 서울_동별 평균 경쟁률')
plt.xlabel('서울_동')
plt.ylabel('평균 경쟁률')
plt.xticks(rotation=90)
plt.show()
```
