```python
import pandas as pd

seoul = pd.read_csv('seoul.csv', encoding='utf-8')
gyeonggi = pd.read_csv('gyeonggi.csv', encoding='utf-8')
incheon = pd.read_csv('incheon.csv', encoding='utf-8')
```

```python
seoul
```

```python
gyeonggi
```

```python
gyeonggi_1 = gyeonggi.drop(['Unnamed: 0'], axis=1)
gyeonggi_1
```

```python
incheon
```

```python
incheon_1 = incheon.drop(['Unnamed: 0'], axis=1)
incheon_1
```

```python
price = pd.concat([seoul, gyeonggi_1, incheon_1], ignore_index=True)
price
```

```python
cols = ['지역'] + [col for col in price.columns if col != '지역']
df = price[cols]
df
```

```python
df.to_csv('청약매물_연도별_거래금액 (서울, 경기, 인천).csv', encoding='utf-8', index=False)
```
