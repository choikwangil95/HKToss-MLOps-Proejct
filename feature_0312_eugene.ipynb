```python
df_parcel_price = pd.read_csv("./src/datasets/청약매물_공급금액 (서울, 경기, 인천).csv", encoding="euc-kr")
df_parcel_price.head()
```

```python
df_parcel_price.columns.unique()
```

```python
df_parcel_price = df_parcel_price[["공고번호", "주택명",'주택형', "공급지역명", "공급금액(최고가 기준)"]]
df_parcel_price.head()
```

```python
df_parcel_price = df_parcel_price.loc[df_parcel_price['공급지역명']=='서울']
df_parcel_price.head()
```

```python
df = pd.merge(df, df_parcel_price, on="공고번호", how="left")
df.head()
```

```python
df = df.drop(columns=['주택명_y', '공급지역명_y'], axis=1)
df.head()
```

```python
df['공고번호'].nunique()
```
