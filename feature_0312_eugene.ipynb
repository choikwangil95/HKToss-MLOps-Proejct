```python
import pandas as pd
import numpy as np

df = pd.read_csv("./src/features/train-0311.csv", encoding="euc-kr")
df.head()
```

```python
df_filtered = df[df['공급지역명'].isin(['서울', '경기', '인천'])]
df_filtered.sort_values(by='공급금액(최고가 기준)', ascending=False, inplace=True)
df_filtered.head()
```

```python
df_filtered.sort_values(by='공급금액(최고가 기준)', ascending=False, inplace=True)
df_filtered.head()
```

```python
df_width = df_filtered.groupby(['주택명', '전용면적', '주택형'])['공급금액(최고가 기준)'].sum() / 4
df_width = pd.DataFrame(df_width)``
df_width.reset_index(inplace=True)
df_width
```
